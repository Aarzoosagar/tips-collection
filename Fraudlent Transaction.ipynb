```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# Assume the dataset is named 'creditcard.csv' and contains columns 'Time', 'Amount', 'Class', and various feature columns
data = pd.read_csv("creditcard.csv (1).zip")
```

```python
# Display the first few rows of the dataset
print(data.head())

# Check for missing values
print(data.isnull().sum())
```

```python
# Distribution of target variable (Class)
plt.figure(figsize=(6, 4))
sns.countplot(data['Class'])
plt.title('Distribution of Target Variable (Class)')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
# Distribution of transaction amounts
plt.figure(figsize=(10, 6))
sns.histplot(data['Amount'], bins=50, kde=True)
plt.title('Distribution of Transaction Amounts')
plt.xlabel('Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), cmap='coolwarm', annot=True, fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
