# QUESTION 8

```python
# for total_bill
values, base = np.histogram(df['total_bill'], bins=40)
cumulative = np.cumsum(values)
plt.plot(base[:-1], cumulative)
plt.title('total_bill',fontsize=15)
plt.grid(axis='y', alpha=0.75)
plt.grid(axis='x', alpha=0.75)
plt.show()
```

```python
# for tip
values, base = np.histogram(df['tip'], bins=40)
cumulative = np.cumsum(values)
plt.plot(base[:-1], cumulative)
plt.title('Tip',fontsize=15)
plt.grid(axis='y', alpha=0.75)
plt.grid(axis='x', alpha=0.75)
plt.show()
```
